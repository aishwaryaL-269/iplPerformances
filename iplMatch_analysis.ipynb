```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
ipl_data = 'ipl_matches_100.csv'
ipl_df = pd.read_csv('ipl_matches_100.csv')
WinningTeams = ipl_df['match_winner'].value_counts().head(5)
team_names = WinningTeams.index
WinCounts = WinningTeams.values
plt.figure(figsize=(12, 7))
plt.bar(team_names, WinCounts, color='Skyblue', edgecolor='black', alpha=0.8)
plt.xlabel("Total Victories", fontsize=12)
plt.title('Top Most Successful Teams', fontsize=14, fontweight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.5)
same_winners = (ipl_df['toss_winner'] == ipl_df['match_winner'])
toss_outcomes = np.where(same_winners, 'Match Won by Toss Winner', 'Match Lost by Toss Winner')
labels, values = np.unique(toss_outcomes, return_counts=True)
plt.figure(figsize=(8, 8))
plt.pie(values, labels=labels, autopct='%1.1f%%', 
        colors=['teal', 'Orange'], startangle=140, explode=(0.05, 0))
plt.title("Winning the Toss and the Match", fontsize=13)
top_franchise = team_names[0]
toss_winPercent = (values[labels == 'Match Won by Toss Winner'][0] / values.sum()) * 100
venue_wins = ipl_df['venue'].value_counts().head(10)
plt.figure(figsize=(10, 6))
plt.bar(venue_wins.index, venue_wins.values, color='lavender', edgecolor='black')
plt.title('Top 10 Venues by Matches', fontsize=14, fontweight='bold')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
top_venue = venue_wins.index[0]

print(f"--- Key Insights ---")
print(f"1. Overall Dominance: {top_franchise} has secured the highest number of wins.")
print(f"2. Top Venue: {top_venue} hosted {venue_wins.values[0]} matches.")
print(f"3. Strategic Influence: Winning the toss resulted in a match victory {toss_winPercent:.2f}% of the time.")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as npy

df = pd.read_csv('ipl_matches_100.csv')
top_5 = df['match_winner'].value_counts().head(5)
squads = top_5.index
victories = top_5.values

plot.figure(figsize=(10, 6))
plot.bar(squads, victories, color='salmon', edgecolor='darkred')
plot.ylabel("Total Wins")
plot.title('Top Successful Teams', fontsize=15)
plot.xticks(rotation=15)

m_t_win = (df['toss_winner'] == df['match_winner'])
outcomes = npy.where(m_t_win, 'Won Toss & Match', 'Won Toss, Lost Match')
unique_labels, count_values = npy.unique(outcomes, return_counts=True)

plot.figure(figsize=(7, 7))
plot.pie(count_values, labels=unique_labels, autopct='%1.1f%%', 
        colors=['Orange', 'teal'], startangle=90, explode=(0.05, 0))
plot.title("Relation: Toss vs Match Result")

Top_team = squads[0]
toss_win_rate = (count_values[unique_labels == 'Won Toss & Match'][0] / count_values.sum()) * 100

stadiums = df['venue'].value_counts().iloc[:10]
plot.figure(figsize=(12, 6))
plot.barh(stadiums.index, stadiums.values, color='lightblue')
plot.title('Most Frequent Match Locations')
plot.xlabel('Occurrences')

print(f"--- Key Insights ---")
print(f"Dominant Franchise: {Top_team} has secured the highest number of wins.")
print(f"Popular Stadium: {stadiums.index[0]} hosted {stadiums.values[0]} games")
print(f"Strategic Influence: Winning the toss resulted in a match victory {toss_win_rate:.1f}% of the time.")

plot.show()
```
