```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as npy

df = pd.read_csv('ipl_matches_100.csv')
top_5 = df['match_winner'].value_counts().head(5)
squads = top_5.index
victories = top_5.values

plot.figure(figsize=(10, 6))
plot.bar(squads, victories, color='salmon', edgecolor='darkred')
plot.ylabel("Total Wins")
plot.title('Top Successful Teams', fontsize=15)
plot.xticks(rotation=15)

m_t_win = (df['toss_winner'] == df['match_winner'])
outcomes = npy.where(m_t_win, 'Won Toss & Match', 'Won Toss, Lost Match')
unique_labels, count_values = npy.unique(outcomes, return_counts=True)

plot.figure(figsize=(7, 7))
plot.pie(count_values, labels=unique_labels, autopct='%1.1f%%', 
        colors=['Orange', 'teal'], startangle=90, explode=(0.05, 0))
plot.title("Relation: Toss vs Match Result")

Top_team = squads[0]
toss_win_rate = (count_values[unique_labels == 'Won Toss & Match'][0] / count_values.sum()) * 100

stadiums = df['venue'].value_counts().iloc[:10]
plot.figure(figsize=(12, 6))
plot.barh(stadiums.index, stadiums.values, color='lightblue')
plot.title('Most Frequent Match Locations')
plot.xlabel('Occurrences')

print(f"--- Key Insights ---")
print(f"Dominant Franchise: {Top_team} has secured the highest number of wins.")
print(f"Popular Stadium: {stadiums.index[0]} hosted {stadiums.values[0]} games")
print(f"Strategic Influence: Winning the toss resulted in a match victory {toss_win_rate:.1f}% of the time.")

plot.show()
```
